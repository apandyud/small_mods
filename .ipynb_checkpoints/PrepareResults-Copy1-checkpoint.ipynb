```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas as pd
import os
from pathlib import Path

def load_all_csvs_with_filename_column(directory):
    """
    Betölti a megadott könyvtár összes .csv fájlját egy közös DataFrame-be,
    és hozzáadja a fájlnevet egy 'filename' oszlopban.

    Args:
        directory (str or Path): A könyvtár elérési útja.

    Returns:
        pd.DataFrame: Az egyesített adatkeret.
    """
    all_dfs = []
    directory = Path(directory)

    for csv_file in directory.glob("*.csv"):
        df = pd.read_csv(csv_file)
        df['filename'] = csv_file.name
        all_dfs.append(df)

    if all_dfs:
        combined_df = pd.concat(all_dfs, ignore_index=True)
    else:
        combined_df = pd.DataFrame()

    return combined_df

# Példa használat
# Cseréld ki e

os.remove('res/merged.csv')
results = load_all_csvs_with_filename_column('res')
results.to_csv('res/merged.csv')
```

```python
em_true_counts = (
    results.groupby(['qid', 'question', 'derivation'])['em']
    .apply(lambda x: (x == True).sum())
    .reset_index(name='em_true_count')
)

len(em_true_counts)
```

```python
import re

def replace_numbers(texts, replacement="#"):
    """
    Lecseréli a számokat a megadott helyettesítő karakterláncra.

    Args:
        texts (list of str): Kifejezések listája.
        replacement (str): A csereérték (pl. "#" vagy "NUM").

    Returns:
        list of str: A módosított kifejezések listája.
    """
    number_pattern = r"\d+(?:[\.,]\d+)?"
    return [re.sub(number_pattern, replacement, text).replace(" ","").strip() for text in texts]

```

```python

#devdf = pd.read_json('dataset_raw/tatqa_dataset_dev.json')
#res = pd.read_csv('res/merged.csv')

derivates = list(em_true_counts['derivation'])
derivates = replace_numbers(derivates)
derivates[1:10]

results['der_scheme'] = replace_numbers(results['derivation'])

results['em_true_count'] = (
    results.groupby(['qid', 'question'])['em']
    .transform(lambda x: (x == True).sum())
)

results['em_success_rate'] = (
    results.groupby(['qid', 'question'])['em']
    .transform(lambda x: (x == True).sum() / len(x))
)

results['vm_true_count'] = (
    results.groupby(['qid', 'question'])['good']
    .transform(lambda x: (x == True).sum())
)

results['vm_success_rate'] = (
    results.groupby(['qid', 'question'])['good']
    .transform(lambda x: (x == True).sum() / len(x))
)


results.to_csv('res/merged.csv')
```

```python
# Pivot: kérdés × modell mátrix, érték: em (bool → int)
pivot_df = results.pivot_table(index=['qid', 'question'], columns='filename', values='em', aggfunc='first')

# Missing értékek kitöltése (pl. ha nem minden modell válaszolt minden kérdésre)
pivot_df = pivot_df.fillna(False).astype(int)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Optimalizált klaszterszám keresése (opcionális)
scores = []
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(pivot_df)
    score = silhouette_score(pivot_df, labels)
    scores.append((k, score))
best_k = max(scores, key=lambda x: x[1])[0]

# Clustering futtatása
kmeans = KMeans(n_clusters=best_k, random_state=42)
pivot_df['cluster'] = kmeans.fit_predict(pivot_df)
```

```python
# pivot_df tartalmazza: qid, question, model1, model2, ..., cluster

# 1. Success rate soronként (modellátlag kérdésenként)
pivot_df['success_rate'] = pivot_df.drop(columns='cluster').mean(axis=1)

# 2. Klaszterek átlagos nehézsége
cluster_difficulty = (
    pivot_df.groupby('cluster')['success_rate']
    .agg(['mean', 'count'])
    .reset_index()
    .sort_values('mean')  # Legnehezebb elöl
)
print(cluster_difficulty)
```

```python
# Címke hozzárendelés logikája (pl. kézzel vagy szabály alapján)
def difficulty_label(rate):
    if rate < 0.3:
        return 'hard'
    elif rate < 0.7:
        return 'medium'
    else:
        return 'easy'

pivot_df['difficulty'] = pivot_df['success_rate'].apply(difficulty_label)
```

```python
import matplotlib.pyplot as plt

# PCA koordináták (pl. coords = pca.fit_transform(...))
for diff in ['easy', 'medium', 'hard']:
    mask = pivot_df['difficulty'] == diff
    plt.scatter(coords[mask, 0], coords[mask, 1], label=diff)

plt.legend()
plt.title("Kérdések nehézség szerint színezve")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
pca = PCA(n_components=2)
coords = pca.fit_transform(pivot_df.drop(columns='cluster'))

plt.figure(figsize=(8, 6))
for cluster_id in sorted(pivot_df['cluster'].unique()):
    points = coords[pivot_df['cluster'] == cluster_id]
    plt.scatter(points[:, 0], points[:, 1], label=f'Cluster {cluster_id}')

plt.legend()
plt.title("Question Clustering Based on Answer Patterns")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# 2. Modell betöltése
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# 3. Embeddingek generálása
embeddings = model.encode(derivates)

# 4. Optimális klaszterszám keresése (Silhouette score alapján)
silhouette_scores = []
range_n_clusters = range(3, 10)
for n_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    labels = kmeans.fit_predict(embeddings)
    score = silhouette_score(embeddings, labels)
    silhouette_scores.append(score)

# Legjobb klaszterszám kiválasztása
optimal_clusters = range_n_clusters[silhouette_scores.index(max(silhouette_scores))]

# 5. KMeans újrafuttatása optimális klaszterszámmal
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
labels = kmeans.fit_predict(embeddings)

# 6. Adatok DataFrame-be
df = pd.DataFrame({
    'derivates': derivates,
    'cluster': labels
})

# 7. Embeddingek 2D-re csökkentése PCA-val vizualizációhoz
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embeddings)

# 8. Vizualizáció
plt.figure(figsize=(10, 6))
for cluster_id in range(optimal_clusters):
    cluster_points = reduced_embeddings[labels == cluster_id]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f"Cluster {cluster_id}")
for i, question in enumerate(derivates):
    plt.annotate(f"{i}", (reduced_embeddings[i, 0], reduced_embeddings[i, 1]))
plt.title(f"Question Clustering (k={optimal_clusters})")
plt.legend()
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.grid(True)
plt.tight_layout()
plt.show()

```
